# Visualizing Distribution of Data
1. distplot()
2. kdeplot()
3. jointplot()

```python
from numpy.random import randn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
x=randn(10000)
```

```python
x
```

```python
#distplot-graph with normal distribution curve
sns.distplot(x)
plt.show()
```

```python
#kdeplot-only curve will be appeared
sns.kdeplot(x)
plt.show()
```

```python
#jointplot -least used in data
tips=sns.load_dataset('tips')
tips.tail()
x=tips['total_bill']
y=tips['tip']
sns.jointplot(x=x,y=y)
```
