# Linear Regression And Relationship
1. regplot()
2. lmplot()

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#regplot-used to see or predict the future growth and result of data
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#lmplot-same as regplot
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```
