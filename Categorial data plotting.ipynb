# Categorial Data Plotting
1. Catplot()
2. Stripplot()
3. Swarmplot()
4. Countplot()

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Catplot-overlapped dotted graph
sns.catplot(x='day', y='total_bill', data=tips)
plt.show()   # it is used to show max or min of particular in data sets
```

```python
#Swarmplot-sorted grapgh but same as catplot
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
#Countplot-bar grapgh with countable data
sns.countplot(x='day', data=tips)
plt.show()
```

```python
plt.figure(figsize=(10,5))   #size of graph or box
plt.grid()   # to show grid lines
sns.countplot(x='size',hue='time',data=tips)   #hue is for colour which helps in differentiate
plt.yticks(range(0,110,10))    #range on y axis
plt.show()
```

```python
#Stripplot  it is same don't have much differnece
sns.stripplot(x='smoker',y='total_bill',data=tips)
plt.show()
```
