# Matplotlib Scatter Matrix Graph

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.plotting import scatter_matrix
```

```python
iris=sns.load_dataset('iris')
df=iris.drop(['species'],axis=1)
df.head()
```

```python
# For kde= kernel density estimation
scatter_matrix(df,figsize=(10,8),color='r',diagonal='kde')
plt.show()      # in this row and column will be compared and if both data are  ot related it will be scattered 
```
