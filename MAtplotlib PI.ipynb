# Matplotlib pi-graph

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rupee_monthly=[2000,2500,900,15000,1500]
item=['Tution_fee','Mobile_bill','Electricity','Rent','Music']

# To plt data
plt.pie(x=rupee_monthly,labels=item)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=rupee_monthly,labels=item,explode=[0.3,0,0,0,0],autopct='%.2f',rotatelabels=True)
# here explode is used to seprate particular and autopct to show and %.2f= 2 decimals it is immutable and initial in explode is 0
```
